```python
import pandas as pd
```

```python
#import file
df_confirmed = pd.read_csv('cases_covid19_confirmed.csv')
df_confirmed
```

```python
df_confirmed1 = df_confirmed.groupby('Country/Region').sum().reset_index()
df_confirmed_cleaned = df_confirmed1.drop(columns=['Province/State', 'Lat', 'Long'])
df_confirmed_cleaned
```

```python
#transposed
df_confirmed_transposed = df_confirmed_cleaned.transpose()
#re-clean
df_confirmed_transposed.columns = df_confirmed_transposed.iloc[0]
df_confirmed_transposed = df_confirmed_transposed[1:]
df_confirmed_transposed = df_confirmed_transposed.reset_index().rename(columns={'index':'Date'})

#issues with index. Will change in excel
df_confirmed_transposed 
```

```python
# saved to file
excel_file_path = 'df_confirmed_transposed.xlsx'
df_confirmed_transposed.to_excel(excel_file_path, index=False)
```

```python
# same process as above but for new file
df_deaths = pd.read_csv('cases_covid19_deaths.csv')
df_deaths
```

```python
df_deaths1 = df_deaths.groupby('Country/Region').sum().reset_index()
df_deaths_cleaned = df_deaths1.drop(columns=['Province/State', 'Lat', 'Long'])
df_deaths_cleaned
```

```python
#transposed
df_deaths_transposed = df_deaths_cleaned.transpose()

#re-clean
df_deaths_transposed.columns = df_deaths_transposed.iloc[0]
df_deaths_transposed = df_deaths_transposed[1:]
df_deaths_transposed = df_deaths_transposed.reset_index().rename(columns={'index':'Date'})

#issues with index. Will change in excel
df_deaths_transposed
```

```python
excel_file_path = 'df_deaths_transposed.xlsx'
df_deaths_transposed.to_excel(excel_file_path, index=False)
```
