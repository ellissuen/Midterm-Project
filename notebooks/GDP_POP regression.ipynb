```python
#open file

df_gdp = pd.read_csv('GDP.csv')
df_gdp
```

```python
#open file

df_pop = pd.read_csv('Population.csv')
df_pop
```

```python
#open file
# renaming column

df_country = pd.read_csv('COVID-19_Country.csv')
df_country1 = df_country.rename(columns={'Country_Region':'Country'})
```

```python
df_country
```

```python
# assuming that we only want to use year 2022
#combining all dataframes

df_all = pd.merge(df_country1.drop(columns=['Lat', 'Long_']),
                    df_gdp.drop(columns=['2019', '2020', '2021', '2023']),
                   on='Country')
df_all = pd.merge(df_all, df_pop.drop(columns=['2019', '2020', '2021', '2023']), on='Country')
```

```python
# more cleaning

df_all = df_all.rename(columns={'2022_x':'GDP', '2022_y':'Population'})
df_all = df_all.dropna()
#pd.set_option('display.max_rows', None)
```

```python
df_all
```

```python
df_all = df_all.dropna(subset=['Confirmed', 'Population'])
```

```python
df_all.shape
```

```python
df_all[numeric_columns].describe()
```

```python
#GDP and Population changed to numeric types

df_all[['GDP', 'Population']] = df_all[['GDP', 'Population']].replace(',', '', regex=True).astype(float)
```

```python
df_all.info()
```

```python
#adding new %____ columns for percent of population

df_all['%Confirmed'] = df_all['Confirmed']/(df_all['Population']*1000000)*100
df_all['%Deaths'] = df_all['Deaths']/(df_all['Population']*1000000)*100
```

```python
df_all
```

```python
subset_df = df_all.iloc[:,1:7]

# Create a pair plot
sns.pairplot(subset_df)
plt.show()
```

```python
#Remove OUTLIERS

import numpy as np
from scipy.stats import zscore

numeric_columns = subset_df.select_dtypes(include=np.number).columns

# Calculate Z-scores for each numeric column
z_scores = np.abs(zscore(subset_df[numeric_columns]))

# Define a threshold for outlier detection (e.g., 3 standard deviations)
threshold = 3

# Create a boolean mask for outliers
outlier_mask = (z_scores > threshold).any(axis=1)

# Remove rows containing outliers
df_all_no_outliers = subset_df[~outlier_mask]

# Create a pair plot without outliers
sns.pairplot(subset_df_no_outliers)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df_all['GDP'], df_all['Population'], df_all['Confirmed'])
ax.set_xlabel('GDP')
ax.set_ylabel('Population')
ax.set_zlabel('Confirmed Cases')

plt.show()
```

```python
# Leverage-Residual Squared Plot
from statsmodels.graphics.regressionplots import plot_leverage_resid2
plot_leverage_resid2(model)
plt.show()
```

```python
# Assuming 'GDP', 'Population', and 'Confirmed' are your independent variables and the dependent variable
X1 = df_all['GDP']
X2 = df_all['Population']
y = df_all['Confirmed']

# Create a meshgrid for 3D plotting
X1, X2 = np.meshgrid(X1, X2)

# Create the design matrix
X_design = np.column_stack((np.ones(X1.size), X1.flatten(), X2.flatten()))

# Calculate the predicted values using the model coefficients
y_pred = np.dot(X_design, model.params)

# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot for the observed data
ax.scatter(X1, X2, y, c='r', marker='o', label='Observed Data')

# Surface plot for the regression model
ax.plot_surface(X1, X2, y_pred.reshape(X1.shape), color='b', alpha=0.3, label='Regression Model')

ax.set_xlabel('GDP')
ax.set_ylabel('Population')
ax.set_zlabel('Confirmed Cases')
ax.set_title('3D Regression Model Visualization')

plt.legend()
plt.show()
```
